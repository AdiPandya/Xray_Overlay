# Overlay Guide
This notebook contains guide for ovelaying X-ray image on the Optical image using FITS liberator and GIMP.
Read the instructions in README.md file before using this guide.

NOTE: Before starting, make sure the FITS liberator and GIMP are installed on your system and you have the aligned X-ray and Optical images.

This guide is inspired from the guide provided by Chandra X-ray Observatory. You can find the original guide [here](https://chandra.harvard.edu/photo/openFITS/).

## Why use FITS liberator?

The overlay is done completely in GIMP, but the main reason to use FITS liberator is to change the scaling of the X-ray image. Usually the X-ray image is in linear scale, and it is difficult to see the details in the image when it is ovelayed on the Optical image. FITS liberator can be used to apply powerlaw or arcsinh scaling to the X-ray image to bring out the details.

The image below shows the layout of the FITS liberator window. In the center is the X-ray image, below it is the histogram of the image, and on the right are the controls for scaling. In the scaling tab, you can change the scaling of the image to powerlaw or arcsinh. There, there are options to change the minimum and maximum pixel values of the image and power-law exponent. 
The histogram tab shows the histogram of the image, and you can change the minimum and maximum pixel values of the image by dragging the sliders below the histogram. Make sure you are not clipping the image.

Sometimes the xray image is flipped when opened in FITS liberator. You can flip it back by clicking on the "Flip Image" option available in the right most tab.

Once you are satisfied with the scaling, you can save the image as a .tif file.

# Using GIMP to overlay X-ray and Optical images

Once you have the X-ray and Optical images, you can use GIMP to overlay them. 
Here is how GIMP window looks like.

On the left are the tools such as select, crop, color picker, text, etc. On the bottom right are the layers. You can add the X-ray and Optical images as layers. In the center is the main window where you can see the images once they are opened as layers.

Here are the steps to overlay the images:
1. Open the X-ray image in GIMP. <br />
   Once the image is opened, go to "Image" -> "Mode" -> "RGB" to convert the image to RGB mode.
2. Add the Optical image as a new layer. <br />
   Go to "File" -> "Open as Layers" and select the Optical image.
3. Change the blending mode of the X-ray layer to "Screen". <br />
   You can change the blending mode by clicking on the drop down menu in the layers tab as shown in the image below.

4. Change the color of the X-ray layer. <br />
   The X-ray image is usually in grey scale. You can change the color of the X-ray layer by clicking on the "Colors" -> "Colorize" and selecting the color you want from the pop-up window. Here is the image of the pop-up window with purple color selected.

5. Adjusting the levels. <br />
   You can adjust the levels of the X-ray layer by clicking on "Colors" -> "Levels". You can adjust the levels to make the X-ray image more visible. The image below shows the pop-up window for levels. There are 3 sliders for adjusting the levels. The left slider is for adjusting the black level, the middle slider is for adjusting the mid level, and the right slider is for adjusting the white level. The histogram of the image is shown above the sliders for reference. Play around with the sliders to get the best result. Click on OK once you are done.<br />
   You can also use the "Curves" option to adjust the levels. The curves option gives more control over the levels. Select the option "Edit these Settings as Curves" to use the curves option.

<strong>IMPORTANT NOTE:</strong> Make a duplicate of the X-ray layer before adjusting the levels. Once you change the levels, it can not be undone.You can do this by right clicking on the X-ray layer and selecting "Duplicate Layer". Adjust the levels of the duplicate layer. This way, you can always go back to the original X-ray layer if you mess up the levels.
